```python
import pandas as pd

# Loading data set

df = pd.read_csv('iris.csv')

print(df.head())

# Filter example

filtered = df[df['sepal_length'] >= 4.9]

summary = filtered.describe()

print(filtered.head(50))
print(summary)

filtered = filtered.reset_index(drop = True) # To reset the index

filtered.to_csv('filtered_output.csv', index=True)
summary.to_csv('summary_output.csv')

```

```python
## Building reusable script

%%writefile pipeline.py
import pandas as pd
import sys

input_file = sys.argv[1]   # get the file path from command line
df = pd.read_csv(input_file)

def run_pipeline(input_path, output_filtered, output_summary):
  df = pd.read_csv(input_path)
  filtered = df[df['sepal_length'] >= 4.9]
  summary = filtered.describe()

  filtered.to_csv(output_filtered, index=False)
  summary.to_csv(output_summary)

if __name__ == "__main__":
  input_path = sys.argv[1]
  run_pipeline(input_path, 'filtered.csv', 'summary.csv')
```
